```python
# To remove MatplotlibDeprecationWarning: The iterable function was deprecated in Matplotlib 3.1 and will be removed in 3.3. Use np.iterable instead.
# if not cb.iterable(width):
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# Importing libraries
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
er = nx.erdos_renyi_graph(100, 0.1)
plt.subplot(1, 2, 1)
nx.draw(er, node_size=10, node_color='red')
plt.subplot(1, 2, 2)
plt.ylabel("number of nodes")
plt.xlabel("degrees")
plt.plot(nx.degree_histogram(er))
print(nx.info(er))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

star = nx.star_graph(9)
plt.subplot(1, 2, 1)
nx.draw(star, node_size=400, node_color='red', with_labels=True, font_weight='bold')
plt.subplot(1, 2, 2)
plt.yticks(np.arange(0, 11))
plt.xticks(np.arange(0, 11))
plt.ylim(top=11)
plt.xlim(right=11)
plt.ylabel("number of nodes")
plt.xlabel("degrees")
plt.plot(nx.degree_histogram(star))
print(nx.info(star))
```

```python
petersen = nx.petersen_graph()
plt.subplot(1, 2, 1)
nx.draw_shell(petersen, nlist=[range(5, 10), range(5)], node_size=200, node_color='red', with_labels=True, font_weight='bold')
plt.subplot(1, 2, 2)
plt.yticks(np.arange(0, 11))
plt.xticks(np.arange(0, 11))
plt.ylim(top=11)
plt.xlim(right=11)
plt.ylabel("number of nodes")
plt.xlabel("degrees")
plt.plot(nx.degree_histogram(petersen))
print(nx.info(petersen))
```

```python
cycle = nx.cycle_graph(10)
plt.subplot(1, 2, 1)
nx.draw_spring(cycle, node_size=400, node_color='red', with_labels=True, font_weight='bold')
plt.subplot(1, 2, 2)
plt.yticks(np.arange(0, 11))
plt.xticks(np.arange(0, 11))
plt.ylim(top=11)
plt.xlim(right=11)
plt.ylabel("number of nodes")
plt.xlabel("degrees")
plt.plot(nx.degree_histogram(cycle))
print(nx.info(cycle))
```
