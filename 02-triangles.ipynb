```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# For networks to be drawn inline
%matplotlib inline

G = nx.karate_club_graph()
plt.figure(figsize=(6, 6))
nx.draw_spring(G, node_size=400, node_color='red', with_labels=True, font_weight='bold')

print("n =", nx.number_of_nodes(G))
print("m =", nx.number_of_edges(G))
A = nx.adjacency_matrix(G).todense()
print(A)
print(A*A)

print("sum", np.sum(A))
print("trace", np.trace(A))
```

```python
# Qn1: Make a program of counting the number of triangles in “karate club network”. Show the code and its results.

# Print dictionary of all triangles found in the karate club network
G_triangles = nx.triangles(G)
print(G_triangles)

# Print number of all triangles in karate club network 
# (note division by 3 as each triangle is counted thrice)
G_triangles_sum = int(sum(G_triangles.values()) / 3)
print("Total number of triangles in the karate club network = ", G_triangles_sum)
```

```python
# Qn2: Make a program of computing the maximum number of triangles in a graph of 9 nodes. Show the code and its results.

# The maximum number of triangles would appear in a complete graph (all nodes interconnected)
H = nx.complete_graph(9)

# Draw the graph to check
nx.draw_spring(H, node_size=400, node_color='red', with_labels=True, font_weight='bold')

# Print dictionary of all triangles found in the karate club network
H_triangles = nx.triangles(H)
print(H_triangles)

# Print max number of triangles in network of 9 nodes 
# (note division by 3 as each triangle is counted thrice)
max_num_triangles = int(sum(H_triangles.values()) / 3)
print("Maximum number of triangles in a network of 9 nodes = ", max_num_triangles)
```

```python
# Make a program of drawing a graph of 9 nodes and 12 edges that contains no triangles. Show the code and its results.

# Create an empty graph and add 9 nodes and corresponding edges
J = nx.Graph()
J.add_nodes_from([1,2,3,4,5,6,7,8,9])
J.add_edges_from([(1,2), (2,3), (3,6), (6,9), (8,9), (7,8), (4,7), (1,4), (2,5), (4,5), (5,6), (5,8)])

# Draw the graph with 9 nodes and 12 edges and no triangles
nx.draw_spring(J, node_size=400, node_color='red', with_labels=True, font_weight='bold')

if sum(nx.triangles(J).values()) == 0:
  print("This network has no triangles!")
```
