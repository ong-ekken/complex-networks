```python
# To remove MatplotlibDeprecationWarning: The iterable function was deprecated in Matplotlib 3.1 and will be removed in 3.3. Use np.iterable instead.
# if not cb.iterable(width):
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

import networkx as nx
import matplotlib.pyplot as plt
from networkx.utils.random_sequence import powerlaw_sequence
```

## For Erdos Renyi Graph

```python
er = nx.erdos_renyi_graph(10000, 0.001)
nx.draw(er, node_size=1, node_color='red')
plt.show()
```

```python
print(nx.info(er))
print("C. Components  :", nx.number_connected_components(er))
print("Triangles      :", int(sum(nx.triangles(er).values()) / 3))
print("Transitivity   :", nx.transitivity(er))
er_deg = dict(er.degree)
print("Max Degree     :", max(er_deg.values()))
print("Min Degree     :", min(er_deg.values()))
```

```python
# Figure Plots

# Figure 1 - Degree Distribution Histogram (Linear)
plt.subplot(121)
plt.plot(nx.degree_histogram(er))

# Figure 2 - Degree Distribution Histogram (Log)
plt.subplot(122)
plt.xscale("log")
plt.yscale("log")
plt.grid(which="both")
plt.plot(nx.degree_histogram(er))

plt.show()
```
