drawing graph of polynomial ( $y=x^3-7x^2+12x$)

```python
import matplotlib.pyplot as plt

x = range(500)
y = [0] * 500
for i in x:
  y[i] = (i/100)**3 - 7*(i/100)**2 + 12*(i/100) # y=x^3-7x^2+12x
plt.plot(list(map(lambda x: x * 0.01, x)), y)
```

graphical solution for the size of the giant component

```python
import matplotlib.pyplot as plt

x = range(100)
y = [0] * 100
plt.plot(list(map(lambda x: x * 0.01, x)), list(map(lambda x: 1-np.exp(-1.5*(x * 0.01)), x)))
plt.plot(list(map(lambda x: x * 0.01, x)), list(map(lambda x: 1-np.exp(-1*(x * 0.01)), x)))
plt.plot(list(map(lambda x: x * 0.01, x)), list(map(lambda x: 1-np.exp(-0.5*(x * 0.01)), x)))
plt.plot(list(map(lambda x: x * 0.01, x)), list(map(lambda x: x * 0.01, x)))
plt.show()
```

quiz10: draw a graph of S
$S=1-e^{-cS}$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

x = range(500)
y = [0] * 500
for i in x:
    y[i] = optimize.newton(lambda x: x-1+np.exp(-(i/100)*x), 1)
plt.plot(list(map(lambda x: x * 0.01, x)), y)
plt.show()
```
